```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import warnings
warnings.filterwarnings(action="ignore")
```

```python
data=pd.read_csv("zomato.csv",encoding="latin-1")
data.head()
```

```python
data.dtypes
```

```python
data.isnull().sum()
```

```python
data["Cuisines"]=data['Cuisines'].fillna('NaN')
```

```python
data.isnull().sum()
```

```python
data1=pd.read_excel("Country_Code.xlsx")
data1.head()
```

```python
fd=pd.merge(data,data1,on="Country Code",how="left")
fd.head()
```

```python
fd.Country.value_counts()
```

```python
country_name=fd.Country.value_counts().index
country_name
```

```python
country_value=fd.Country.value_counts().values
country_value
```

```python
plt.bar(country_name,country_value)
plt.xticks(rotation=90)
plt.show()
```

```python
px.pie(names=country_name,values=country_value,width=700,height=700)
```

```python
sns.barplot(x=fd.Country.value_counts().index,y=fd.Country.value_counts().values)
plt.xticks(rotation=90)
plt.show()
```

```python
px.bar(x=fd.Country.value_counts().index,y=fd.Country.value_counts().values,text=fd.Country.value_counts().values)
```

```python
ratings=fd.groupby(['Aggregate rating','Rating color','Rating text']).size()
ratings
```

```python
fd.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index()
```

```python
fd.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings=fd.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

```python
fig=px.bar(ratings,x="Rating color",color="Rating Count",
           color_discrete_sequence=['White','Red','Orange','Yellow','Green','Darkgreen'],text="Rating Count",
           title="Rating Distribution")
fig.update_layout(bargap=0.1)
fig.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x="Aggregate rating",y="Rating Count", data=ratings)
plt.show
```
